### Types of Neural Networks

Three distinct types of neural networks are commonly employed in machine learning:

1. **Artificial Neural Networks (ANNs):**
   - ANNs constitute the foundational architecture of neural networks. Comprising interconnected nodes resembling the human brain's structure, they are versatile in learning complex relationships and approximating nonlinear functions. ANNs are used for diverse tasks including classification, regression, and more.

2. **Convolutional Neural Networks (CNNs):**
   - CNNs are specifically tailored for processing grid-like data such as images and videos. With convolutional layers, CNNs automatically extract features from images, making them proficient in capturing spatial patterns and establishing hierarchical representations within images.

3. **Recurrent Neural Networks (RNNs):**
   - RNNs excel at processing sequential data where order matters. They incorporate hidden states that retain information from previous steps, making them adept at tasks involving time-series data, text, and sequential information.

**Advantages of Convolutional Neural Networks (CNNs):**

Among these neural network types, Convolutional Neural Networks (CNNs) stand out as particularly suitable for image analysis. The following reasons elucidate why CNNs are exceptionally convenient for tasks involving images:

1. **Automated Feature Extraction:** CNNs possess built-in mechanisms that autonomously learn and uncover relevant image features. Manual feature extraction, commonly needed in traditional approaches, is rendered unnecessary.

2. **Spatial Hierarchy Capturing:** CNNs adeptly capture spatial hierarchies and intricate image elements. This empowers them to discern edges, textures, and complex features, facilitating a nuanced comprehension of relationships within images.

3. **Translation Invariance:** The inherent translation invariance of CNNs enables pattern recognition regardless of feature position in an image. This attribute proves vital in tasks like object detection, accommodating variable object placement.

4. **Pretrained Models Availability:** Pretrained CNN models, trained on vast datasets like ImageNet, are readily accessible. Leveraging these models for specific task fine-tuning expedites progress and optimizes resource usage.

5. **Effective Transfer Learning:** CNNs' ability to transfer learned features between tasks enhances adaptability to various image-related challenges, even when confronted with limited data.

In summary, the automatic feature learning, spatial hierarchy capturing, and translation invariance capabilities of Convolutional Neural Networks position them as a potent choice for diverse visual recognition tasks.

### CNN Architecture

A Convolutional Neural Network (CNN) is a specialized type of neural network designed for processing grid-like data, such as images. Its architecture is inspired by the visual processing mechanism of the human brain. A CNN consists of several layers, each performing specific operations on the input data to progressively extract meaningful features and patterns. The architecture typically includes convolutional layers, pooling layers, and fully connected layers.

| Layer Name              | Description                                                                                                                                                |
|-------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------|
| **Input:**              | The input to a CNN is an image or a grid-like data structure. For example, an image can be represented as a grid of pixel values, where each pixel contains information about color intensity. |
| **Convolutional Layer:** | Convolutional layers are the core building blocks of a CNN. They consist of filters (also known as kernels) that slide over the input image. The filter extracts local features by performing element-wise multiplications and summing the results. This process is known as convolution. Convolutional layers help detect edges, textures, and other low-level features in the image. |
| **Pooling Layer:**       | Pooling layers reduce the spatial dimensions of the data while retaining important features. Max-pooling is a common technique used in CNNs. It involves dividing the input image into non-overlapping regions and selecting the maximum value from each region. Pooling helps reduce the computational load and makes the network more robust to variations in the input. |
| **Fully Connected Layer:** | After the convolutional and pooling layers, the extracted features are flattened and passed through one or more fully connected layers. These layers operate like traditional neural network layers, connecting every neuron to every neuron in the previous and subsequent layers. The fully connected layers help combine extracted features to make high-level decisions. |
| **Output:**             | In a CNN designed for classification tasks, the final output is a probability distribution over classes. The last fully connected layer is typically a softmax layer that assigns probabilities to different classes. Each neuron in the output layer corresponds to a class, and the final output values represent the likelihood of the input belonging to each class. The class with the highest probability is considered the predicted class. |



In essence, a Convolutional Neural Network processes input data through convolutional layers to extract local features, pooling layers to reduce spatial dimensions, and fully connected layers to make high-level decisions. The final output is a probability distribution over classes, indicating the network's prediction for the input's classification among the available classes.

### Image Processing

The size of the input images is a critical factor to consider when designing and training a Neural Network for image classification. Larger images imply a larger number of pixels; therefore, more parameters and computational resources are required to process and train the model. 

The size of the input images may also affect the model's ability to generalize to new images and its ability to detect relevant features at different scales.

Reducing the size of images in your dataset for a neural network classifier can offer several benefits and trade-offs, depending on your specific use case and computational constraints. Here are some reasons why you might consider resizing your images:

1. **Memory and Computational Efficiency:** Larger images require more memory and computational resources to process during training and inference. Resizing images to a smaller resolution reduces the memory footprint of your model, allowing you to train and deploy your neural network more efficiently.

2. **Faster Training:** Smaller images generally lead to faster training times. Processing smaller images requires fewer calculations, resulting in quicker iterations during training. This is especially important when experimenting with different model architectures, hyperparameters, or data augmentation techniques.

3. **Model Complexity:** Smaller images can allow you to use larger or more complex neural network architectures without worrying as much about computational constraints. This can potentially improve the model's capacity to learn complex features from the data.

4. **Overfitting Mitigation:** Training with large images can sometimes increase the risk of overfitting, where the model memorizes the training data rather than generalizing well to new data. Smaller images may help mitigate overfitting by reducing the model's capacity to memorize fine details in the training data.

5. **Noise Reduction:** Smaller images can help reduce noise and irrelevant details present in high-resolution images. This can lead to a more focused model that learns relevant features for classification.

6. **Data Augmentation:** When applying data augmentation techniques (such as rotation, cropping, and flipping) to increase the diversity of your dataset, working with smaller images can be more manageable and result in faster data augmentation during training.

However, it's important to consider the potential trade-offs when reducing image size:

- **Loss of Detail:** Smaller images can result in a loss of fine details and may impact the model's ability to recognize intricate patterns.
  
- **Reduced Performance for Certain Tasks:** For tasks where high-resolution details are crucial, such as object detection or image segmentation, reducing image size might lead to suboptimal performance.

- **Contextual Information:** Smaller images might lose contextual information that could be important for certain classifications.

Ultimately, the decision to resize images should be based on a balance between computational constraints, the specific requirements of the task, and the trade-offs between model performance and efficiency.

Another approach is to change the images from **3D** to **2D**. In the context of neural networks and deep learning, converting an image from 3D to 2D refers to the process of flattening the spatial dimensions of the image while preserving the information contained in the channels or color components. 

An image is typically represented as a 3D array, where the dimensions correspond to height, width, and channels (e.g., Red, Green, and Blue channels in a color image). When converting from 3D to 2D, the spatial dimensions (height and width) are collapsed into a single dimension, resulting in a long vector. The channels, on the other hand, remain separate.

This transformation is often necessary when transitioning from convolutional layers, which operate on 3D image data with spatial relationships, to fully connected (dense) layers that require flattened input. The flattened 2D vector serves as the input to these dense layers.

The process of converting from 3D to 2D can be visualized as unstacking the rows of pixels in each channel and arranging them in a single line, while keeping each channel's information distinct. This transformation allows neural networks to process the image data using fully connected layers, which are common in the final stages of many neural network architectures.

In summary, converting an image from 3D to 2D in the context of neural networks involves flattening the spatial dimensions while preserving the channel information, facilitating the transition from convolutional layers to fully connected layers.

There are 3 different datasets that need to be created:

**Training Set:**
The training set is a crucial component of training a neural network. It's a subset of the available data that is used to teach the neural network how to learn patterns and relationships within the data. During training, the neural network adjusts its weights and biases based on the input data and the corresponding target outputs. This iterative process helps the network learn to make accurate predictions or classifications by minimizing the difference between predicted and actual outputs.

**Testing Set:**
The testing set is used to assess the performance of a trained neural network. It consists of data that the network has never seen during the training process. The network's predictions on the testing set provide an indication of how well it generalizes to new, unseen data. Evaluating the network's accuracy, precision, recall, and other performance metrics on the testing set helps determine how well it is likely to perform in real-world scenarios.

**Validation Set:**
The validation set serves as an intermediary step between the training and testing phases. It is used during the training process to monitor the network's performance on data that it hasn't been explicitly trained on. By evaluating the network's performance on the validation set, adjustments to hyperparameters, network architecture, or training methods can be made to enhance its performance before final testing. The validation set helps prevent overfitting, where a neural network memorizes the training data instead of learning generalizable patterns.

In summary, the training set teaches the neural network patterns in the data, the validation set aids in fine-tuning the model during training, and the testing set provides an objective assessment of the network's performance on new, unseen data. These different sets play vital roles in building and evaluating the efficacy of neural network models.

### Neural Network Architecture Configuration and Training

The code defines the architecture of a Convolutional Neural Network (CNN) using the Keras framework. The model is constructed sequentially using the `Sequential` class from Keras. It is designed to process images of specific dimensions (img_width, img_height, 3), where the last value '3' signifies the number of color channels (e.g., red, green, blue) in the input images.

The CNN architecture is as follows:

1. A `Conv2D` layer with 10 filters of size (3, 3) is defined as the initial layer. The 'same' padding ensures the output has the same spatial dimensions as the input. Batch normalization is applied to normalize the output of the convolutional layer. The ReLU activation function introduces non-linearity. A MaxPooling2D layer with a pool size of (2, 2) is used for downsampling.

2. The process is repeated with additional convolutional layers to progressively extract more complex features. There are layers with 20, 30, 40, and 50 filters, each with batch normalization, ReLU activation, and max-pooling applied afterward.

3. Following the convolutional layers, a `Flatten` layer is introduced to convert the 2D feature maps into a 1D vector, preparing them for fully connected layers.

4. The final layer is a `Dense` layer with 10 neurons, representing the output classes. It employs the softmax activation function to produce a probability distribution over the classes, allowing the network to assign a probability to each class label.

In summary, this code segment establishes a CNN architecture composed of multiple convolutional layers with batch normalization and ReLU activation, followed by max-pooling layers for feature extraction. The fully connected layers and softmax activation in the final layer allow the network to make predictions across the specified classes. This architecture is designed to learn and recognize patterns in the input image data for classification tasks.

```python
# Define the CNN architecture
model = Sequential()
model = models.Sequential([
    layers.Conv2D(10, (3, 3), padding='same', input_shape=(img_width, img_height, 3)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(20, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(30, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(40, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(50, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(10, activation='softmax')
])
```

In the graph above we can see how the RMSProp shows good results, however, the SGD shows greater repeatability in the results, which is why this was selected.

```python
# Parameters
num_runs = 5
learning_rate = 0.04
momentum = 0.9

# Lists to store validation accuracy values for each algorithm
val_accuracy_results_rmsprop = []
val_accuracy_results_sgd = []

# Loop through the runs for RMSPROP
for run in range(num_runs):
    print(f"Running RMSPROP - Run {run + 1}...")
    
    # Define the CNN architecture
    model = Sequential()
    model = models.Sequential([
    layers.Conv2D(10, (3, 3), padding='same', input_shape=(img_width, img_height, 3)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(20, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(30, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(40, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(50, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(10, activation='softmax')
])
    
    # Create an instance of the optimizer
    optimizer = RMSprop(learning_rate=learning_rate)
    
    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit(train_generator,
                        validation_data=test_generator,
                        epochs=150,
                        verbose=0)
    
    val_accuracy_results_rmsprop.append(history.history['val_accuracy'][-1])

# Loop through the runs for SGD
for run in range(num_runs):
    print(f"Running SGD - Run {run + 1}...")
    
    ## Define the CNN architecture
    model = Sequential()
    model = models.Sequential([
    layers.Conv2D(10, (3, 3), padding='same', input_shape=(img_width, img_height, 3)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(20, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(30, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(40, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(50, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(10, activation='softmax')
])
    
    # Create an instance of the optimizer
    optimizer = SGD(learning_rate=learning_rate, momentum=momentum)
    
    model.compile(optimizer=optimizer,
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    
    history = model.fit(train_generator,
                        validation_data=test_generator,
                        epochs=150,
                        verbose=0)
    
    val_accuracy_results_sgd.append(history.history['val_accuracy'][-1])

# Plot the validation accuracy results
plt.figure(figsize=(10, 6))

plt.plot(range(1, num_runs + 1), val_accuracy_results_rmsprop, 'o-', label='RMSPROP')
plt.plot(range(1, num_runs + 1), val_accuracy_results_sgd, 's-', label='SGD')

plt.xlabel('Run Number')
plt.ylabel('Final Validation Accuracy')
plt.title('Final Validation Accuracy Comparison')
plt.legend()
plt.xticks(np.arange(1, num_runs + 1))
plt.show()
```

```python
# Rest of the code...

# Learning rate values to iterate through
learning_rates = [0.02, 0.04, 0.06, 0.08, 0.1, 0.15, 0.2]
num_runs = 5

# List to store final validation accuracy values for each learning rate and run
final_val_accuracy_results = []

# Loop through the learning rates
for lr in learning_rates:
    print(f"Running with learning rate: {lr}")
    
    run_val_accuracy = []
    
    for run in range(num_runs):
        print(f"    Run: {run + 1}")
        # Define the CNN architecture
        model = Sequential()
        model = models.Sequential([
        layers.Conv2D(10, (3, 3), padding='same', input_shape=(img_width, img_height, 3)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(20, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(30, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(40, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Conv2D(50, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        layers.Flatten(),
        layers.Dense(10, activation='softmax')
        ])
        # Define the optimizer with the current learning rate
        optimizer = SGD(learning_rate=lr, momentum=momentum)
        
        # Compile the model
        model.compile(optimizer=optimizer,
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
        
        # Train the model
        history = model.fit(train_generator,
                            validation_data=test_generator,
                            epochs=150,
                            verbose=0)
        
        # Save the final validation accuracy value for this run
        run_val_accuracy.append(history.history['val_accuracy'][-1])
    
    # Save the final validation accuracy values for all runs of this learning rate
    final_val_accuracy_results.append(run_val_accuracy)

# Plot the final validation accuracy results for different learning rates
plt.figure(figsize=(10, 6))

for i, lr in enumerate(learning_rates):
    plt.plot(range(1, num_runs + 1), final_val_accuracy_results[i], 'o-', label=f"LR = {lr}")

plt.xlabel('Run Number')
plt.ylabel('Final Validation Accuracy')
plt.title('Final Validation Accuracy vs Run Number')
plt.legend()
plt.xticks(np.arange(1, num_runs + 1))
plt.show()
```
